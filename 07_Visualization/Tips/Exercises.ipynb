### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv('https://raw.githubusercontent.com/guipsamora/'
                   + 'pandas_exercises/master/07_Visualization/Tips/tips.csv')
```

```python
tips.head()
```

### Step 4. Delete the Unnamed 0 column

```python
tips = tips.drop(columns = 'Unnamed: 0')
```

```python
tips.head()
```

### Step 5. Plot the total_bill column histogram

```python
fig, ax = plt.subplots()

sns.distplot(tips['total_bill'], kde=False, ax=ax)
ax.set(xlabel='Total Bill', ylabel='Frequency', title='Total Bill')
sns.despine()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
fig, ax = plt.subplots()

sns.scatterplot(x='total_bill', y='tip', data=tips, ax=ax)
ax.set(xlabel='Total Bill', ylabel='Tips', title='Total Bill by Tips')
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
fig, ax = plt.subplots()

sns.scatterplot(x='total_bill', y='tip', data=tips, hue='size')
ax.set(xlabel='Total Bill', ylabel='Tips', title='Total Bill by Tips, separated by Size')
```

```python
sns.pairplot(tips)
```

### Step 8. Present the relationship between days and total_bill value

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
fig, ax = plt.subplots()

sns.stripplot(y='day', x='tip', data=tips, hue='sex', jitter=True)
ax.set(xlabel='Tips', ylabel='Day', title='Tips by Day, Separated ')
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
fig, ax = plt.subplots()

sns.boxplot(y='day', x='total_bill', data=tips, hue='time')
ax.set(xlabel='Total Bill', ylabel='Day', title='Total Bill by Day, Separated by Time')
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
facet = sns.FacetGrid(tips, col='time')
facet.map(plt.hist, 'tip')
```

```python
fig, axes = plt.subplots(ncols=2, sharey=True)

sns.distplot(tips.loc[tips['time'] == 'Dinner', 'tip'], kde=False, ax =axes[0])
sns.distplot(tips.loc[tips['time'] == 'Lunch', 'tip'], kde=False, ax =axes[1])
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
facet = sns.FacetGrid(tips, col='sex', hue='smoker')
facet.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')

facet.add_legend()
```
